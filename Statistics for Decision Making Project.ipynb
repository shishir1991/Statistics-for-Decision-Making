#### 1. For the suburb Altona, it is postulated that a typical property sells for 800,000 dollars. Use the data at hand to test this assumption. Is the typical property price really 800,000 dollars or has it increased? Use a significance level of 5%.

```python
from scipy.stats import ttest_1samp


# Filter the dataset for properties in Altona
altona_data = df[df['Suburb'] == 'Altona']

# Extract the prices of the properties
altona_prices = altona_data['Price'].dropna()

# Perform the one-sample t-test
test_result = ttest_1samp(altona_prices, 800000, alternative='greater')

# Extract the test statistic and p-value
test_statistic = test_result.statistic
p_value = test_result.pvalue

# Count the number of properties in Altona
n_properties = len(altona_prices)

print(f"Test Statistic: {test_statistic}")
print(f"P-value: {p_value}")
print(f"Number of Properties: {n_properties}")

# Based on the significance level of 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The mean property price in Altona is greater than $800,000.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude that the mean property price in Altona is greater than $800,000.")
```

#### 2. For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and rest as summer months. Use a significance level of 5%. 

```python
from scipy.stats import ttest_ind

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Filter the dataset for the year 2016
data_2016 = df[df['Date'].dt.year == 2016]

# Define winter and summer months
winter_months = [1, 2, 3, 10, 11, 12]
summer_months = [4, 5, 6, 7, 8, 9]

# Extract property prices for winter and summer months
winter_prices = data_2016[data_2016['Date'].dt.month.isin(winter_months)]['Price'].dropna()
summer_prices = data_2016[data_2016['Date'].dt.month.isin(summer_months)]['Price'].dropna()

# Perform the independent samples t-test
test_statistic, p_value = ttest_ind(winter_prices, summer_prices, equal_var=False)

# Print the results
print(f"Test Statistic: {test_statistic}")
print(f"P-value: {p_value}")

# Based on the significance level of 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in mean property prices between summer and winter months.")
else:
    print("Fail to reject the null hypothesis: No significant difference in mean property prices between summer and winter months.")
```

#### 3. For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places.

```python
from scipy.stats import binom

# Filter the dataset for the suburb of Abbotsford
abbotsford_data = df[df['Suburb'] == 'Abbotsford']

# Calculate the probability of a property not having car parking in Abbotsford
p_no_parking = 1 - (abbotsford_data['Car'] > 0).mean()

# Number of trials
n = 10

# Number of successes we are interested in
k = 3

# Calculate the probability using the binomial probability formula
probability = binom.pmf(k, n, p_no_parking)

# Round off the answer to 3 decimal places
probability_rounded = round(probability, 3)

# Print the result
print(f"The probability that out of 10 properties sold in Abbotsford, 3 will not have car parking is {probability_rounded}.")
```

#### 4. In the suburb Abbotsford, what are the chances of finding a property with 3 rooms? Round your answer to 3 decimal places. 

```python
# Filter the dataset for properties in Abbotsford with 3 rooms
abbotsford_3_rooms =df[(df['Suburb'] == 'Abbotsford') & (df['Rooms'] == 3)]

# Calculate the chances of finding a property with 3 rooms in Abbotsford
chances_3_rooms = len(abbotsford_3_rooms) / len(df[df['Suburb'] == 'Abbotsford'])

# Print the chances rounded to 3 decimal places
print(f"Chances of finding a property with 3 rooms in Abbotsford: {chances_3_rooms:.3f}")
```
